```python
import torch
import os
import sys
import numpy as np


project_root = os.path.abspath('.') 
if project_root not in sys.path:
    sys.path.append(project_root)

from experiments.nns_benchmark import run_nns_benchmark
from experiments.lm_benchmark import run_lm_benchmark

DEVICE_STR = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Running on: {DEVICE_STR}")
print("Ensure 'xlstm_replica.py' is in the project root directory.")
print("Ensure the official 'xlstm' library is installed.")
```

```python
print("\n==========================================================")
print("STARTING NNS BENCHMARK 1: Scratch xLSTM vs. Library xLSTM")
print("==========================================================")
nns_results_1 = run_nns_benchmark(device_str=DEVICE_STR, benchmark_type="xlstm_vs_lib")
print("\nNNS Benchmark 1 - Final MSE Results:")
if nns_results_1:
    for model_name, mse in nns_results_1.items():
        print(f"{model_name}: {mse:.6f}")
else:
    print("NNS Benchmark 1 did not produce results.")
```

```python
print("\n=====================================================================")
print("STARTING NNS BENCHMARK 2: Baselines (LSTM, Transformer) vs. xLSTM")
print("=====================================================================")
nns_results_2 = run_nns_benchmark(device_str=DEVICE_STR, benchmark_type="baselines_vs_xlstm")
print("\nNNS Benchmark 2 - Final MSE Results:")
if nns_results_2:
    for model_name, mse in nns_results_2.items():
        print(f"{model_name}: {mse:.6f}")
else:
    print("NNS Benchmark 2 did not produce results.")
```

```python
print("\n========================================================")
print("STARTING LM BENCHMARK 1: Scratch xLSTM vs. Library xLSTM")
print("========================================================")
lm_results_1 = run_lm_benchmark(device_str=DEVICE_STR, benchmark_type="xlstm_vs_lib")
print("\nLM Benchmark 1 - Final PPL Results (Overall):")
if lm_results_1:
    for model_name, ppl_list in lm_results_1.items():
        overall_ppl = ppl_list[-1] if ppl_list and len(ppl_list) > 0 else float('nan')
        print(f"{model_name}: {overall_ppl:.2f}" if np.isfinite(overall_ppl) else f"{model_name}: N/A")
else:
    print("LM Benchmark 1 did not produce results.")
```

```python
print("\n==================================================================")
print("STARTING LM BENCHMARK 2: Baselines (LSTM, Transformer) vs. xLSTM")
print("==================================================================")
lm_results_2 = run_lm_benchmark(device_str=DEVICE_STR, benchmark_type="baselines_vs_xlstm")
print("\nLM Benchmark 2 - Final PPL Results (Overall):")
if lm_results_2:
    for model_name, ppl_list in lm_results_2.items():
        overall_ppl = ppl_list[-1] if ppl_list and len(ppl_list) > 0 else float('nan')
        print(f"{model_name}: {overall_ppl:.2f}" if np.isfinite(overall_ppl) else f"{model_name}: N/A")
else:
    print("LM Benchmark 2 did not produce results.")

print("\n===================================================")
print("ALL BENCHMARKS COMPLETE")
print("===================================================")
```
